# Cómo construir los modelos de Markov

Este notebook es un esqueleto: cada celda tiene instrucciones para completar los pasos.

Para crear los modelos se usan estas etapas:

1. Tokenizar el texto en palabras limpias.
2. Construir las transiciones según el orden del modelo:
   - Orden 1: `transiciones[palabra_actual].append(palabra_siguiente)`.
   - Orden 2: `transiciones[(palabra_a, palabra_b)].append(palabra_siguiente)`.
   - Orden 3: `transiciones[(palabra_a, palabra_b, palabra_c)].append(palabra_siguiente)`.
3. Calcular las probabilidades de cada palabra siguiente dividiendo el conteo de cada candidato por el total de ocurrencias del estado.
4. Generar texto eligiendo la siguiente palabra según la distribución de probabilidades y avanzando el estado.
