```python
# Celda 1: Cargar el texto
import os

# Ruta al archivo de texto
ruta_texto = 'textos/crimen_y_castigo.txt'

# Leer el archivo
with open(ruta_texto, 'r', encoding='utf-8') as f:
    texto = f.read()

print(f"Texto cargado. Longitud: {len(texto)} caracteres.")
print("Primeras 500 caracteres:")
print(texto[:500])
```

```python
# Celda 2: Preprocesamiento básico (Fácil)
import re
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

# Limpiar texto: remover puntuación y convertir a minúsculas
texto_limpio = re.sub(r'[^\w\s]', '', texto.lower())

# Tokenizar
palabras = word_tokenize(texto_limpio, language='spanish')

print(f"Número de palabras: {len(palabras)}")
print("Primeras 20 palabras:", palabras[:20])

# Pregunta 1: ¿Cuál es la palabra más frecuente? (Usa collections.Counter)
```

```python
# Celda 3: Modelo de Markov Orden 1 (Intermedio)
from collections import defaultdict, Counter

# Construir transiciones: diccionario de listas
transiciones = defaultdict(list)
for i in range(len(palabras) - 1):
    transiciones[palabras[i]].append(palabras[i+1])

# Calcular probabilidades
probabilidades = {}
for palabra, siguientes in transiciones.items():
    total = len(siguientes)
    probs = Counter(siguientes)
    probabilidades[palabra] = {sig: count / total for sig, count in probs.items()}

print("Ejemplo de transiciones para 'el':", probabilidades.get('el', 'No encontrado'))

# Pregunta 2: Completa la frase 'el amor' usando orden 1. ¿Cuál es la palabra más probable siguiente?
```

```python
# Celda 4: Modelo de Markov Orden 2 (Avanzado)
# Transiciones para pares de palabras
transiciones_2 = defaultdict(list)
for i in range(len(palabras) - 2):
    par = (palabras[i], palabras[i+1])
    transiciones_2[par].append(palabras[i+2])

probabilidades_2 = {}
for par, siguientes in transiciones_2.items():
    total = len(siguientes)
    probs = Counter(siguientes)
    probabilidades_2[par] = {sig: count / total for sig, count in probs.items()}

print("Ejemplo para ('el', 'amor'):", probabilidades_2.get(('el', 'amor'), 'No encontrado'))

# Pregunta 3: Completa 'el amor' usando orden 2. Genera 3 palabras siguientes.
```

```python
# Celda 5: Modelo de Markov Orden 3 y Generación (Avanzado)
import random

# Transiciones para triples
transiciones_3 = defaultdict(list)
for i in range(len(palabras) - 3):
    triple = (palabras[i], palabras[i+1], palabras[i+2])
    transiciones_3[triple].append(palabras[i+3])

probabilidades_3 = {}
for triple, siguientes in transiciones_3.items():
    total = len(siguientes)
    probs = Counter(siguientes)
    probabilidades_3[triple] = {sig: count / total for sig, count in probs.items()}

# Función para generar texto
def generar_texto(inicio, orden, longitud=10):
    if orden == 1:
        probs = probabilidades
        estado = inicio[-1]
    elif orden == 2:
        probs = probabilidades_2
        estado = tuple(inicio[-2:])
    elif orden == 3:
        probs = probabilidades_3
        estado = tuple(inicio[-3:])
    
    texto_generado = list(inicio)
    for _ in range(longitud):
        if estado not in probs:
            break
        siguiente = random.choices(list(probs[estado].keys()), weights=list(probs[estado].values()))[0]
        texto_generado.append(siguiente)
        if orden == 1:
            estado = siguiente
        else:
            estado = tuple(texto_generado[-orden:])
    return ' '.join(texto_generado)

# Ejemplo
inicio = ['el', 'amor']
print("Generado orden 1:", generar_texto(inicio, 1, 5))
print("Generado orden 2:", generar_texto(inicio, 2, 5))
print("Generado orden 3:", generar_texto(inicio + ['es'], 3, 5))  # Asumiendo 'es' como tercera

# Pregunta 4: Completa 'el amor' hasta 20 palabras usando orden 3. ¿Qué observas sobre la coherencia?
```
